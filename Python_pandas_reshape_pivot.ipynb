## Pandas Reshape and Pivot

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.pivot_table(index='day',columns="sex")
```

```python
pd.pivot_table(tips,index="day",columns='sex',values='tip',aggfunc=np.average)
```

```python
pd.pivot_table(tips,index=["day","time"],columns='sex',values='tip',aggfunc=[np.average,np.sum])
```

```python
pd.pivot_table(tips,index=["day","time"]
               ,columns='sex'
               ,values='tip'
               ,aggfunc=[np.average,np.sum]
              ,margins=True)
```
